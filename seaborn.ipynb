**Que 1: Name any five plots that we can plot using the Seaborn library. Also, state the uses of each plot.**

Sol 1:
Seaborn is a powerful Python data visualization library built on top of Matplotlib. It provides a high-level interface for creating attractive and informative statistical graphics. Here are five types of plots that you can create using Seaborn, along with their common uses:

Scatter Plot:

Use: A scatter plot is used to visualize the relationship between two numerical variables. Each data point is represented by a dot, and the position of the dot on the x and y axes corresponds to the values of the two variables.
Common Uses: To explore correlations, clusters, and trends between two variables.
Line Plot:

Use: A line plot is used to display the trend of a numerical variable over time or another continuous dimension. It connects data points with lines, making it useful for visualizing changes in data over a continuous interval.
Common Uses: Tracking trends in data, visualizing time series data, and showing patterns or changes in data over time.
Bar Plot:

Use: A bar plot (bar chart) is used to display and compare the frequency or distribution of categorical data. The height of each bar represents the count or proportion of data in each category.
Common Uses: Showing comparisons between different categories, visualizing frequency distributions, and displaying categorical data.
Histogram:

Use: A histogram is used to visualize the distribution of a numerical variable. It groups the data into bins or intervals and displays the frequency or density of data points in each bin.
Common Uses: Understanding the underlying distribution of data, identifying data skewness, and analyzing data spread.
Heatmap:

Use: A heatmap is used to visualize the relationships between two categorical variables by using color intensity. It displays a matrix of colors where each cell's color represents the value of the relationship between two variables.
Common Uses: Analyzing the correlation between multiple categorical variables, visualizing confusion matrices, and displaying matrices with color-coded values.

**Que 2: Load the "fmri" dataset using the load_dataset function of seaborn. Plot a line plot using x = "timepoint" and y = "signal" for different events and regions.**

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Load the "fmri" dataset
fmri_data = sns.load_dataset("fmri")

# Create a line plot using Seaborn
plt.figure(figsize=(10, 6))
sns.lineplot(data=fmri_data, x="timepoint", y="signal", hue="event", style="region")

# Add title and labels
plt.title("FMRI Signal Over Time by Event and Region")
plt.xlabel("Timepoint")
plt.ylabel("Signal")

# Display the plot
plt.legend()
plt.show()
```
