## Q1: What is Matplotlib? Why is it used? Name five plots that can be plotted using the Pyplot module of Matplotlib.

Solution -
Matplotlib is a popular Python library used for creating static, interactive, and animated visualizations in a wide range of formats. It provides a flexible and comprehensive set of tools for creating various types of plots and graphs, making it a fundamental tool for data visualization in the Python ecosystem.

Matplotlib is used to create visualizations that help convey information and insights from data. It is commonly used in data analysis, scientific research, engineering, and more. Some of the key features of Matplotlib include support for a variety of plot types, customization of plot appearance, integration with NumPy arrays, LaTeX rendering for mathematical expressions, and support for various output formats like images, interactive displays, and more.

Five plots that can be created using the Pyplot module of Matplotlib are:

- **Line Plot:** A basic plot that displays data points connected by straight lines. It's often used to visualize trends over time or to compare different datasets.
- **Bar Plot:** A plot that represents data using rectangular bars with lengths proportional to the values they represent. Bar plots are used to compare data across different categories.
- **Scatter Plot:** A plot that displays individual data points as dots in a two-dimensional space. Scatter plots are used to visualize the relationship between two continuous variables.
- **Histogram:** A graphical representation of the distribution of numerical data. It consists of a series of bins, where data points are grouped based on their values.
- **Pie Chart:** A circular chart divided into sectors, each representing a portion of the whole. Pie charts are used to display the proportion of different categories within a dataset.

## Q2: What is a scatter plot? Use the following code to generate data for x and y. Using this generated data plot a scatter plot.

**Scatter Plot:** A plot that displays individual data points as dots in a two-dimensional space. Scatter plots are used to visualize the relationship between two continuous variables.

```python
import numpy as np
np.random.seed(3)
x = 3 + np.random.normal(0, 2, 50)
y = 3 + np.random.normal(0, 2, len(x))
```

```python
import matplotlib.pyplot as plt
```

```python
plt.scatter(x,y)
plt.title("Plot of x against y")
plt.xlabel("X label")
plt.ylabel("Y label")
```

## Q3: Why is the subplot() function used? Draw four line plots using the subplot() function.

The subplot() function in Matplotlib is used to create multiple plots within a single figure, arranged in a grid. This is useful for comparing multiple datasets or visualizing different aspects of the same data side by side. Each subplot can have its own set of axes and can display different types of plots.

```python
import numpy as np
import matplotlib.pyplot as plt

# Data
x = np.array([0, 1, 2, 3, 4, 5])

# Define y values for each line
y1 = np.array([0, 100, 200, 300, 400, 500])
y2 = np.array([50, 20, 40, 20, 60, 70])
y3 = np.array([10, 20, 30, 40, 50, 60])
y4 = np.array([200, 350, 250, 550, 450, 150])

# Create a 2x2 grid of subplots
plt.subplot(2, 2, 1)  # First subplot
plt.plot(x, y1, label='Line 1')
plt.title('Line Plot 1')
plt.legend()

plt.subplot(2, 2, 2)  # Second subplot
plt.plot(x, y2, label='Line 2')
plt.title('Line Plot 2')
plt.legend()

plt.subplot(2, 2, 3)  # Third subplot
plt.plot(x, y3, label='Line 3')
plt.title('Line Plot 3')
plt.legend()

plt.subplot(2, 2, 4)  # Fourth subplot
plt.plot(x, y4, label='Line 4')
plt.title('Line Plot 4')
plt.legend()

# Adjust layout to prevent overlapping of subplots
plt.tight_layout()

# Show the plots
plt.show()

```

## Q5: What is a box plot? Why is it used? Using the following data plot a box plot.

A box plot, also known as a box-and-whisker plot, is a graphical representation of the distribution of a dataset. It displays key statistical measures and characteristics of the data in a concise and visual manner. A box plot consists of a rectangular "box" that represents the interquartile range (IQR) of the data, along with "whiskers" that extend from the box to indicate the range of the data.

Here's how a box plot is constructed:

- The box represents the IQR, which is the range between the first quartile (25th percentile) and the third quartile (75th percentile) of the data. The width of the box is typically uniform.
- The median (50th percentile) is indicated by a line or another smaller box inside the main box.
- The whiskers extend from the edges of the box to the minimum and maximum values within a certain distance from the box. The distance can be determined based on certain rules, like 1.5 times the IQR.
- Data points that fall outside the whiskers are considered outliers and are often plotted individually.

Box plots are used for several purposes:

- **Summary of Distribution:** Box plots provide a visual summary of the central tendency (median) and spread (IQR) of the data. They allow you to quickly understand the distribution without having to compute extensive statistical measures.
- **Comparison:** Box plots are useful for comparing the distribution of multiple datasets. By placing several box plots side by side, you can easily compare their medians, ranges, and variability.
- **Identification of Outliers:** Box plots help identify potential outliers in the data by highlighting values that fall outside the whiskers. This can be valuable for detecting anomalies or extreme values.
- **Skewness and Symmetry:** The shape of the box plot can indicate the skewness of the distribution. For example, if one whisker is much longer than the other, it suggests skewness.
- **Data Distribution:** Box plots provide information about the overall distribution of the data, such as whether it's symmetric, skewed, or bimodal.
- **Robustness:** Box plots are robust to outliers and extreme values, making them a useful tool when dealing with datasets that might have irregularities.
Overall, box plots are an essential tool in exploratory data analysis

```python
box1 = np.random.normal(100, 10, 200)
box2 = np.random.normal(90, 20, 200)
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Combine the data into a list for plotting
data = [box1, box2]

# Create a box plot
plt.boxplot(data, labels=['Box 1', 'Box 2'])
plt.title('Box Plot')
plt.xlabel('Boxes')
plt.ylabel('Values')

# Show the plot
plt.show()
```
